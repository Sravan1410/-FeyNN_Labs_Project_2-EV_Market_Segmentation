```python
# STEP 3: Count of manufacturers per state
state_counts = df_spec['State'].value_counts().reset_index()
state_counts.columns = ['State', 'No. of EV Manufacturers']

print("Manufacturers per state:")
display(state_counts)
```

```python
from matplotlib import pyplot as plt
state_counts['No. of EV Manufacturers'].plot(kind='hist', bins=20, title='No. of EV Manufacturers')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
state_counts['No. of EV Manufacturers'].plot(kind='line', figsize=(8, 4), title='No. of EV Manufacturers')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# STEP 4: Visualization
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.barplot(data=state_counts, x='No. of EV Manufacturers', y='State', hue='State', palette='Blues_d', legend=False)
plt.title('EV Manufacturer Count by State')
plt.xlabel('Number of Manufacturers')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```
